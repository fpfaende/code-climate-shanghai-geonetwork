### download a shanghai extract
We use osmnx to downlaod shanghai graph

We use a custom square 

```python
jingan = Polygon([(left,top),(right,top),(right,bottom),(left,bottom)])
```

```python
G = ox.graph_from_polygon(jingan, network_type='drive')
```

### display a graph
display the map using osmnx 

```python
fig, ax = ox.plot_graph(G, edge_linewidth=0.3)
```
